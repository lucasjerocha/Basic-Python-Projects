```python
# A program that allows calculation of areas and volumes of various solids and figures.

import math

def calculate_area_rectangle():
    base = float(input("Enter the base length: "))
    height = float(input("Enter the height: "))
    area = base * height
    print("The area of the rectangle is:", area)

def calculate_area_triangle():
    base = float(input("Enter the base length: "))
    height = float(input("Enter the height: "))
    area = 0.5 * base * height
    print("The area of the triangle is:", area)

def calculate_area_pentagon():
    perimeter = float(input("Enter the perimeter: "))
    apothem = float(input("Enter the apothem length: "))
    area = 0.5 * perimeter * apothem
    print("The area of the pentagon is:", area)

def calculate_area_circle():
    radius = float(input("Enter the radius: "))
    area = math.pi * radius ** 2
    print("The area of the circle is:", area)

def calculate_volume_sphere():
    radius = float(input("Enter the radius: "))
    volume = (4/3) * math.pi * radius ** 3
    print("The volume of the sphere is:", volume)

def calculate_volume_cone():
    radius = float(input("Enter the radius: "))
    height = float(input("Enter the height: "))
    volume = (1/3) * math.pi * radius ** 2 * height
    print("The volume of the cone is:", volume)

def calculate_volume_cylinder():
    radius = float(input("Enter the radius: "))
    height = float(input("Enter the height: "))
    volume = math.pi * radius ** 2 * height
    print("The volume of the cylinder is:", volume)

def calculate_volume_parallelepiped():
    width = float(input("Enter the width: "))
    length = float(input("Enter the length: "))
    height = float(input("Enter the height: "))
    volume = width * length * height
    print("The volume of the parallelepiped is:", volume)

def main():
    option = int(input("To calculate area, enter 1; to calculate volume, enter 2: "))
    if option == 1:
        shape = int(input("Enter the shape (1-Rectangle, 2-Triangle, 3-Pentagon, 4-Circle): "))
        if shape == 1:
            calculate_area_rectangle()
        elif shape == 2:
            calculate_area_triangle()
        elif shape == 3:
            calculate_area_pentagon()
        elif shape == 4:
            calculate_area_circle()
        else:
            print("Invalid shape selection.")
    elif option == 2:
        solid = int(input("Enter the solid (1-Sphere, 2-Cone, 3-Cylinder, 4-Parallelepiped): "))
        if solid == 1:
            calculate_volume_sphere()
        elif solid == 2:
            calculate_volume_cone()
        elif solid == 3:
            calculate_volume_cylinder()
        elif solid == 4:
            calculate_volume_parallelepiped()
        else:
            print("Invalid solid selection.")
    else:
        print("Invalid option.")

if __name__ == "__main__":
    main()
```
