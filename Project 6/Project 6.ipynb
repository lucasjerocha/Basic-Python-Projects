```python
import csv
import numpy as np
import pylab
import math

#111111111111111111111111111111111111111
print("\n1)", 70 * "-", "\n\n")
def csv_to_matrix1(filename):
    with open(filename, newline="", encoding='utf-8') as l:
        matrix = np.loadtxt(l, dtype=str, delimiter=",")
        return matrix

def csv_to_matrix2(filename):
    with open(filename, newline="", encoding='utf-8') as l:
        matrix = np.loadtxt(l, dtype=str, delimiter=",")
        return matrix

matrix1 = csv_to_matrix1("/Users/lujerocha/Desktop/file0.csv") # you must put your own file path
print(matrix1)
print("\n")
matrix2 = csv_to_matrix2("/Users/lujerocha/Desktop/file9.csv") # you must put your own file path
print(matrix2)

#22222222222222222222222222222222222222222
print("\n\n2)", 70 * "-", "\n\n")
def merge_matrices():
    empty = np.empty_like(matrix1)
    empty[:] = matrix1
    unique_matrix = np.append(empty, np.zeros([len(empty), 1]), 1)
    for i in range(len(matrix1)):
        for j in range(len(matrix2)):
            if matrix1[i][0] == matrix2[j][0]:
                unique_matrix[i][6] = matrix2[j][1]
    unique_matrix[0][6] = "Number of words"
    for k in range(1, len(matrix1)):
        if unique_matrix[k][6] == '0.0':
            num_words = int(input("Enter the number of words for the book " + str(unique_matrix[k][1]) + " :"))
            unique_matrix[k][6] = num_words
    return unique_matrix

unique_matrix = merge_matrices()
print(unique_matrix)

f = open("/Users/lucasjerocha/Desktop/file8.csv", 'w', newline='', encoding='utf-8') # you must put your own file path
writer = csv.writer(f)
writer.writerows(unique_matrix)
f.close()

#333333333333333333333333333333333333333333
print("\n\n3)", 70 * "-", "\n\n")
null_matrix = np.zeros([1, 4], dtype=object)
null_matrix[0][0] = 'Genres'
null_matrix[0][1] = 'Maximum'
null_matrix[0][2] = 'Minimum'
null_matrix[0][3] = 'Average'
for i in unique_matrix[1:, 3]:
    if i not in null_matrix[:, 0]:
        max_words = 0
        min_words = math.inf
        sum = 0
        count = 0
        a = 1
        genre_columns = np.array(unique_matrix[:, 3])
        for word in unique_matrix[1:, 6]:
            if genre_columns[a] == i:
                value = int(word)
                if value < min_words:
                    min_words = value
                if value > max_words:
                    max_words = value
                sum += int(word)
                count += 1
            a += 1
        average = sum / count
        null_matrix = np.append(null_matrix, [[i, max_words, min_words, average]], 0)
print(null_matrix)

t = open("/Users/lucasjerocha/Desktop/file7.csv", 'w', newline='') # you must put your own file path
writer = csv.writer(t)
writer.writerows(null_matrix)
t.close()

#444444444444444444444444444444444444444444
print("\n\n4)", 70 * "-", "\n\n")
def plot_graph(w):
    p = np.sort(w)
    x = p
    pylab.hist(x, color="g", bins=15, align="mid")
    pylab.title("Number of books published each year")
    pylab.xlabel("Years")
    pylab.ylabel("Books published")
    pylab.grid()
    pylab.show()

w = matrix1[1:, 5]
plot_graph(w)
```
