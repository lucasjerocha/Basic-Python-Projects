```python
### Exercise 1
### A program that allows entering the statistical data of this variable
### (the salaries of different employees of the company) and that in the end presents a small report
#intro
# Function to calculate the average salary
def calculate_average(salaries):
    return sum(salaries) / len(salaries)

# Function to calculate the total number of employees
def calculate_total(employees):
    return sum(employees)

# Function to find the mode salary
def find_mode(employees):
    max_salary = max(employees)
    mode = employees.index(max_salary) + 1
    return mode, max_salary

def main():
    # Introduction
    print("There are 5 salary ranges in this company:")
    print("- Range 1 - 1000 Euros;")
    print("- Range 2 - 1500 Euros;")
    print("- Range 3 - 2000 Euros;")
    print("- Range 4 - 2500 Euros;")
    print("- Range 5 - 3000 Euros.")
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")

    # Input number of employees in each salary range
    salaries = [1000, 1500, 2000, 2500, 3000]
    employees = []
    for i in range(5):
        while True:
            try:
                num_employees = int(input(f"Enter the number of employees in Range {i+1}: "))
                if num_employees < 0:
                    raise ValueError("Please enter a positive number.")
                employees.append(num_employees)
                break
            except ValueError as e:
                print(e)

    # Generate reports
    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
    print("Average salary of employees: ", calculate_average([salary * employees[i] for i, salary in enumerate(salaries)]))
    print("Total number of employees: ", calculate_total(employees))

    mode, max_salary = find_mode(employees)
    print("Mode salary: Range", mode, "-", max_salary, "Employees")

    print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
    print("Absolute and relative frequency table:")
    print("____________________________________________________")
    print("Salary\t\tAbsolute Freq.\t\tRelative Freq.")
    for i, salary in enumerate(salaries):
        print(salary, "\t\t", employees[i], "\t\t\t", employees[i] / calculate_total(employees))

if __name__ == "__main__":
    main()
```
